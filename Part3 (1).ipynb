```python
from PIL import Image

def rotate_image(image_path, angle):
  try:
    img = Image.open(image_path)
    rotated_img = img.rotate(angle)  # Rotate by the specified angle
    return rotated_img
  except IOError:
    return None

image_path = '/content/1.jpg'

rotated_image_45 = rotate_image(image_path, 45)
rotated_image_90 = rotate_image(image_path, 90)

# Save the rotated images
if rotated_image_45:
  rotated_image_45.save('rotated_45.jpg')
if rotated_image_90:
  rotated_image_90.save('rotated_90.jpg')

print("Rotation done successfully.")

```
