```python
from PIL import Image


def reduce_intensity_levels(image_path, level):

    # Checking if the entered number is power of 2
    if not (level & (level - 1) == 0):
        raise ValueError("num_levels must be a power of 2")

    # Loading the image
    img = Image.open(image_path).convert('L')

    # Getting image data
    image_data = list(img.getdata())

    # Reducing the intensity levels
    divisor = 256 // level
    reduced_data = [pixel // divisor * divisor for pixel in image_data]

    # Create the result as a new image
    new_img = img.copy()
    new_img.putdata(reduced_data)

    return new_img

try:
  level = int(input("Enter the intensity level ( should be power of 2): "))
  new_image = reduce_intensity_levels('/content/1.jpg', level)
  new_image.save('new_image.jpg')
  print("Image saved with", level, "intensity levels.")
except ValueError as e:
  print(e)

```
